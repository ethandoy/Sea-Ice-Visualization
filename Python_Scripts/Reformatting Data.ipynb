```python
import pandas as pd
```

```python
#Load data into Datafram, removing the first row
allData = pd.read_csv("/content/N_seaice_extent_daily_v3.0.csv", skiprows=[1])
allData
```

```python
#clean up column names
allData.rename(str.strip, axis='columns', inplace=True)
allData.columns
```

```python
#remove the missing and source data colums
allData.drop(allData.columns[[4, 5]], axis=1, inplace=True)
allData = allData[allData.Year != 1978]
allData
```

Build a subset of the data with only rows from 2021

```python
# export CSV of all data
allData.to_csv('allData.csv', header=False, index=False)
```

```python
#export CSV with the average for each month of each year
this_data_list = list()
for year in range(1979, 2022):
  for month in range(1, 13):
    mean_extent = allData.loc[(allData['Year']) == year][(allData['Month'] == month)]['Extent'].mean() # find mean for specific month
    this_data_list.append([year, month, mean_extent])    

mean_df = pd.DataFrame(this_data_list) #load list into Dataframe to export as CSV
mean_df.to_csv('meanData.csv', header=False, index=False)

mean_df
```
